```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


uri = "https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/master/aula4.1/movies.csv"

filmes = pd.read_csv (uri) #DataFrame
filmes.head ()
```

```python
import pandas as pd
uri = "https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/master/aula1.2/ratings.csv"
notas = pd.read_csv (uri)
notas.head()
```

```python
notas.columns = ["usuarioID" , "filmesID" , "nota" , "momento"] #Momento seria quando foi votado o filme
notas.head()
```

```python
notas ["nota"].head() #Series
```

```python
notas ["nota"].unique() #Valores nicos, ou seja, as minhas notas dos filmes variam de 4 a 0.5
```

```python
notas.describe() #Resumo estatstico(Media, Moda, Mediana, Desvio Padro...)
```

```python
mediana = notas['nota'].median() #A mediana  representada na tabela acima com 50%
print("\nMediana:", mediana)

moda = notas['nota'].mode()[0] #A moda  o valor que mais se repete, e  representada na tabela com 75%
print("Moda:", moda)

media = notas['nota'].mean() #A mdia  a soma de todos os valores dividido pela quantidade de valores,e  representada com mean
print("Mdia:", media)
```

```python
#Esse grafico mostra a distribuio da minhas notas

notas.nota.plot(kind = 'hist',  color='pink', edgecolor='white')
plt.ylabel('Frequncia')
plt.title('Histograma das Notas')
plt.show()
```

```python
frequencia = notas['nota'].value_counts()
frequencia.head()
```

```python
#O grfico de barras mostra a frequncia de cada nota.

frequencia.plot(kind='bar', color='purple', edgecolor='black', figsize=(8, 6))
plt.title('Frequncia das Notas')
plt.xlabel('Notas')
plt.ylabel('Frequncia')
plt.grid(axis='y')
plt.show()
```

```python
sns.boxplot(x='nota', data=notas, color='purple')
plt.title('Boxplot das Notas')
plt.show()
```

```python
# Calculando a assimetria
assimetria = notas['nota'].skew()
print(f"Assimetria da coluna nota: {assimetria}")
```

A maioria das avaliaes no dataset provavelmente est mais prxima de valores altos, como 4 ou 5, com menos notas prximas de 1 ou 2.


```python
plt.figure(figsize=(8, 7))
sns.histplot(notas['nota'], kde=True, color='Red')
plt.title('Distribuio das Notas')
plt.xlabel('Notas')
plt.ylabel('Frequncia')
plt.grid(True)
plt.show()
```

```python
# Calculando a curtose
curtose = notas['nota'].kurtosis()
print(f"Curtose: {curtose}")
```

Como o valor da curtose  levemente positivo (0.1233), a distribuio  prxima de uma distribuio normal, mas com um pequeno indicativo de que ela tem caudas um pouco mais pesadas e um pico mais acentuado em comparao  distribuio normal.
No entanto, o valor  muito prximo de 0, sugerindo que a distribuio das sugerindo que a distribuio das notas est quase normal, com pouca diferena em termos de caudas ou pico.

```python
plt.figure(figsize=(8, 6))
sns.histplot(notas['nota'], kde=True, color='red')
plt.title('Distribuio das Notas')
plt.xlabel('Nota')
plt.ylabel('Frequncia')
plt.grid(True)
plt.show()
```

```python
generos_separados = filmes['generos'].str.split('|')

todos_generos = [genero for lista_generos in generos_separados for genero in lista_generos]
frequencia_generos = pd.Series(todos_generos).value_counts()

print(frequencia_generos)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=frequencia_generos.values, y=frequencia_generos.index, color = 'purple')
plt.title('Frequncia dos Gneros dos Filmes')
plt.xlabel('Nmero de Filmes')
plt.ylabel('Gnero')
plt.show()
```

```python
top_generos = frequencia_generos.nlargest(10)

plt.figure(figsize=(10, 6))

sns.barplot(x=top_generos.values, y=top_generos.index, color = 'purple')
plt.title('Top 10 dos Filmes')
plt.xlabel('Nmero de Filmes')
plt.ylabel('Gnero')
plt.show()
```
