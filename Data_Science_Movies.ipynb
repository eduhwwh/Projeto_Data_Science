```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


uri = "https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/master/aula4.1/movies.csv"

filmes = pd.read_csv (uri) #DataFrame
filmes.head ()
```

```python
import pandas as pd
uri = "https://raw.githubusercontent.com/alura-cursos/introducao-a-data-science/master/aula1.2/ratings.csv"
notas = pd.read_csv (uri)
notas.head()
```

```python
notas.columns = ["usuarioID" , "filmesID" , "nota" , "momento"] #Momento seria quando foi votado o filme
notas.head()
```

```python
notas ["nota"].head() #Series
```

```python
notas ["nota"].unique() #Valores únicos, ou seja, as minhas notas dos filmes variam de 4 a 0.5
```

```python
notas.describe() #Resumo estatístico(Media, Moda, Mediana, Desvio Padrão...)
```

```python
mediana = notas['nota'].median() #A mediana é representada na tabela acima com 50%
print("\nMediana:", mediana)

moda = notas['nota'].mode()[0] #A moda é o valor que mais se repete, e é representada na tabela com 75%
print("Moda:", moda)

media = notas['nota'].mean() #A média é a soma de todos os valores dividido pela quantidade de valores,e é representada com mean
print("Média:", media)
```

```python
#Esse grafico mostra a distribuição da minhas notas

notas.nota.plot(kind = 'hist',  color='pink', edgecolor='white')
plt.ylabel('Frequência')
plt.title('Histograma das Notas')
plt.show()
```

```python
frequencia = notas['nota'].value_counts()
frequencia.head()
```

```python
#O gráfico de barras mostra a frequência de cada nota.

frequencia.plot(kind='bar', color='purple', edgecolor='black', figsize=(8, 6))
plt.title('Frequência das Notas')
plt.xlabel('Notas')
plt.ylabel('Frequência')
plt.grid(axis='y')
plt.show()
```

```python
sns.boxplot(x='nota', data=notas, color='purple')
plt.title('Boxplot das Notas')
plt.show()
```

```python
# Calculando a assimetria
assimetria = notas['nota'].skew()
print(f"Assimetria da coluna nota: {assimetria}")
```

A maioria das avaliações no dataset provavelmente está mais próxima de valores altos, como 4 ou 5, com menos notas próximas de 1 ou 2.


```python
plt.figure(figsize=(8, 7))
sns.histplot(notas['nota'], kde=True, color='Red')
plt.title('Distribuição das Notas')
plt.xlabel('Notas')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

```python
# Calculando a curtose
curtose = notas['nota'].kurtosis()
print(f"Curtose: {curtose}")
```

Como o valor da curtose é levemente positivo (0.1233), a distribuição é próxima de uma distribuição normal, mas com um pequeno indicativo de que ela tem caudas um pouco mais pesadas e um pico mais acentuado em comparação à distribuição normal.
No entanto, o valor é muito próximo de 0, sugerindo que a distribuição das sugerindo que a distribuição das notas está quase normal, com pouca diferença em termos de caudas ou pico.

```python
plt.figure(figsize=(8, 6))
sns.histplot(notas['nota'], kde=True, color='red')
plt.title('Distribuição das Notas')
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

```python
generos_separados = filmes['generos'].str.split('|')

todos_generos = [genero for lista_generos in generos_separados for genero in lista_generos]
frequencia_generos = pd.Series(todos_generos).value_counts()

print(frequencia_generos)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=frequencia_generos.values, y=frequencia_generos.index, color = 'purple')
plt.title('Frequência dos Gêneros dos Filmes')
plt.xlabel('Número de Filmes')
plt.ylabel('Gênero')
plt.show()
```

```python
top_generos = frequencia_generos.nlargest(10)

plt.figure(figsize=(10, 6))

sns.barplot(x=top_generos.values, y=top_generos.index, color = 'purple')
plt.title('Top 10 dos Filmes')
plt.xlabel('Número de Filmes')
plt.ylabel('Gênero')
plt.show()
```
